```python
# num = 5 # Integer value
print(type(num))
```

```python
num = 10.2  # Float Value
print(type(num))
```

```python
num = 2 + 4j  # Complex Number
print(type(num))
```

```python
print (3**2)
print (9//2) # Floor Division
print (5%2)
```

```python
num = 1
num  += 1
print(num)

num -= 1
print(num)
```

```python
print(3 * (2 + 1)) # 9
print(3 * 2 + 1) # 7
```

```python
# Absolute Value
print(abs(-4)) # O/P : 4

# Round
print(round(4.6345)) # O/P : 5.0

# How many digits we want to round
print(round(1.543532, 3)) # O/P : 1.544
```

```python
# Conditional Operator
num_1 = 3
num_2 = 2
print(num_1 == num_2)
print(num_1 >= num_2)
print(num_1 <= num_2)
```

```python
num_1 = '100'
num_2  = '200'
print(num_1 + num_2)
```

```python
# Typecasting
num = "100"
num_1 = int(num)

print(type(num))
print(type(num_1))
```
