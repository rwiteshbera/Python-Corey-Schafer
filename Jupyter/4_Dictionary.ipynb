```python
# Creating a dictionary
Dict = {1: 'Geeks', 2: 'For', 3: 'Geeks'}
print(Dict)
```

```python
# Dictionary containing mixed keys
Dict1 = {'Name': 'Rwitesh', 1: [1,2,3,4,5]}
print(Dict1)
```

```python
# Creating a nested Dictionary
nested_Dict = {1: 'Geeks', 2: 'For', 
                3: {'A': 'Welcome'}, 'B': 'To', 'C': 'Geeks'}
print(nested_Dict)
```

```python
# Access data from Dictionary
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}
print(Dict['name'])
print(Dict['courses'])
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Add new key-value
Dict['phone'] = 555555
print(Dict.get('phone'))
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Update a key-value pair
Dict.update({'name': 'Rahul'})
print(Dict['name'])
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Delete a specific key
del Dict['age']
print(Dict)
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Print only keys
print(Dict.keys())

# Print only values
print(Dict.values())
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Print keys using loop
for key in Dict:
    print(key)
```

```python
Dict = {'name' : 'Rwitesh', 'age': 20, 'courses': ['Math', 'Physics']}

# Print keys & values using loop
for keys, values in Dict.items():
    print(keys, values)
```
