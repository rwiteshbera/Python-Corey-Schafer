```python
nums = [1, 2, 3, 4, 5]

# For Loop
for e in nums:
    print(e)
```

```python
nums = [1, 2, 3, 4, 5]

# Use of Break
for e in nums:
    if e == 3:
        print('Found!')
        break
    print(e)
```

```python
nums = [1, 2, 3, 4, 5]

# Use of Continue
for e in nums:
    if e == 3:
        print('Continue!')
        break
    print(e)
```

```python
# Nested Loops
nums = [1,2,3,4,5]

for num in nums:
    for letter in 'abc':
        print(num, letter)
```

```python
# Print upto 9
for i in range(10):
    print(i)
```

```python
for i in range(5,15):
    print(i)
```

```python
# While Loop
x = 0

while x < 10:
    print(x)
    x += 1
```
