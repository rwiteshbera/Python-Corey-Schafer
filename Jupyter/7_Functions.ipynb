```python
def evenOdd(x):
    if(x & 1 == 1):
        print("Odd")
    else:
        print("Even")
        
evenOdd(5)
evenOdd(6)
```

```python
def addition(x, y):
    return x + y

print(addition(5,8))
```

```python
def getLength(x):
    return len(x)

print(getLength("Rwitesh"))
```

```python
def greet(text):
    return 'Welcome {}'.format(text)

print(greet("Rwitesh"))
```

```python
def student (*args, **kwargs):
    print(args)
    print(kwargs)
    
student('Math', 'Art', name = 'Rwitesh', age=20)
```

```python
def student (*args, **kwargs):
    print(args)
    print(kwargs)

courses = ['Math', 'Art']
info = {'name': 'Rwitesh', 'age': 20} # Dictionary

student(*courses, **info)
```
